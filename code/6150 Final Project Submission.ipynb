```python
#Import  python libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Importing the dataset
data = pd.read_csv('/Users/taniabadran/Desktop/6150 final project /copy of starbucks stocks .csv')

#Removing the "$" from the columns, and turning the data from a string to a float
data['High'] = data['High'].replace('[\$,]', '', regex=True).astype(float)
data['Low'] = data['Low'].replace('[\$,]', '', regex=True).astype(float)
data['Open'] = data['Open'].replace('[\$,]', '', regex=True).astype(float)
data['Close/Last'] = data['Close/Last'].replace('[\$,]', '', regex=True).astype(float)

#https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html
# Ensure the 'Date' column is in datetime format
data['Date'] = pd.to_datetime(data['Date'])


#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html
# Sorting data just in case it's not in date order
data.sort_values('Date', inplace=True)
```

```python
#Find the highest high price
alltimehigh = data['High'].max()
#Find the lowest high price. 
lowesthighprice = data['High'].min()
#Find the dates: 
alltimehighdate = data[data['High'] == alltimehigh]['Date'].iloc[0]
lowesthighpricedate = data[data['High'] == lowesthighprice]['Date'].iloc[0]
#Printing the results:
print("The lowest high price for Starbucks Company is: $%s (on %s)" % (lowesthighprice, lowesthighpricedate))
print("The highest high price for Starbucks Company is: $%s (on %s)" % (alltimehigh, alltimehighdate))
```

```python
#Find the lowest low price
alltimelow = data['Low'].min()
#Find the highest low price. 
highlowprice = data['Low'].max()
#Find the dates: 
alltimelowdate = data[data['Low'] == alltimelow]['Date'].iloc[0]
highlowpricedate = data[data['Low'] == highlowprice]['Date'].iloc[0]
#Printing the results:
print("The all time low stock price for Starbucks Comapny is: $%s (on %s)"% (alltimelow,alltimelowdate))
print("The highest low price for Starbucks Comapny is: $%s (on %s)"% (highlowprice,highlowpricedate))
```

```python
#Lowest open price
lowopen = data['Open'].min()
#Highest open price. 
highopen = data['Open'].max()
#Find the dates: 
lowopendate = data[data['Open'] == lowopen]['Date'].iloc[0]
highopendate = data[data['Open'] == highopen]['Date'].iloc[0]
#Printing the results:
print("The lowest open price for Starbucks Comapny is: $%s (on %s)"% (lowopen,lowopendate))
print("The highest open price for Starbucks Comapny is: $%s (on %s)"% (highopen,highopendate))
```

```python
#Lowest close price
lowclose = data['Close/Last'].min()
#Highest close price. 
highclose = data['Close/Last'].max()
#Find the dates: 
lowclosedate = data[data['Close/Last'] == lowclose]['Date'].iloc[0]
highclosedate = data[data['Close/Last'] == highclose]['Date'].iloc[0]

#Printing the results:
print("The lowest close price for Starbucks Comapny is: $%s (on %s)"% (lowclose,lowclosedate))
print("The highest close price for Starbucks Comapny is: $%s (on %s)"% (highclose,highclosedate))
```

```python
print("The lowest high price for Starbucks Company is: $%s (on %s)" % (lowesthighprice, lowesthighpricedate))
print("The highest high price for Starbucks Company is: $%s (on %s)" % (alltimehigh, alltimehighdate))

print("The all time low stock price for Starbucks Comapny is: $%s (on %s)"% (alltimelow,alltimelowdate))
print("The highest low price for Starbucks Comapny is: $%s (on %s)"% (highlowprice,highlowpricedate))

print("The lowest open price for Starbucks Comapny is: $%s (on %s)"% (lowopen,lowopendate))
print("The highest open price for Starbucks Comapny is: $%s (on %s)"% (highopen,highopendate))

print("The lowest close price for Starbucks Comapny is: $%s (on %s)"% (lowclose,lowclosedate))
print("The highest close price for Starbucks Comapny is: $%s (on %s)"% (highclose,highclosedate))
```

```python
#https://matplotlib.org/stable/tutorials/pyplot.html
# Plotting the trends of the high and low prices, there isnt any dramatic changes. 
plt.figure(figsize=(10, 6))  
plt.plot(data['Date'], data['High'], marker='', color='blue', linewidth=2, label='High')
plt.plot(data['Date'], data['Low'], marker='', color='red', linewidth=2, label='Low')
plt.title('High and Low Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)  
plt.legend()
plt.tight_layout()  
plt.show()
```

```python
# Plotting the open and close prices, there isnt any dramatic changes.
#The same bc when u open and close its the same
plt.figure(figsize=(10, 6))  # Adjust the size as needed
plt.plot(data['Date'], data['Open'], marker='', color='blue', linewidth=2, label='High')
plt.plot(data['Date'], data['Close/Last'], marker='', color='red', linewidth=2, label='Low')
plt.title('Open VS Close Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)  
plt.legend()
plt.tight_layout()  
plt.show()
```

```python
#https://www.geeksforgeeks.org/pandas-series-dt-year/
# Get the year from the 'Date' column
data['Year'] = data['Date'].dt.year
# Calculate the high average price for each year
highavgperyr = data.groupby('Year')['High'].mean()
print(highavgperyr)
```

```python
# Calculate the low average price for each year
lowavgperyr = data.groupby('Year')['Low'].mean()
print(lowavgperyr)
```

```python
# Calculate the open average price for each year
openavgperyr = data.groupby('Year')['Open'].mean()
print(openavgperyr)
```

```python
# Calculate the close average price for each year
closeavgperyr = data.groupby('Year')['Close/Last'].mean()
print(closeavgperyr)
```

```python
#https://pandas.pydata.org/docs/reference/api/pandas.Series.reset_index.html
# make this a DataFrame
t1 = highavgperyr.reset_index()
t1.columns = ['Year', 'Average High']
print(t1)
```

```python
# make this a DataFrame
t2 = lowavgperyr.reset_index()
t2.columns = ['Year', 'Average Low']
print(t2)
```

```python
# make this a DataFrame
t3 = openavgperyr.reset_index()
t3.columns = ['Year', 'Average Open']
print(t3)
```

```python
# make this a DataFrame
t4 = closeavgperyr.reset_index()
t4.columns = ['Year', 'Average Close']
print(t4)
```

```python
#now merge the data to make a big table 
c1 = pd.merge(t1, t2, on='Year', how='outer')
c2 = pd.merge(c1, t3, on='Year', how='outer')
ft1 = pd.merge(c2, t4, on='Year', how='outer')
print(ft1)
```

```python
#obtain month and year from the date data
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
#join the month and year and then get the average
monthavgyear = data.groupby(['Year', 'Month'])['Close/Last'].mean().reset_index()
# Rename columns
monthavgyear.columns = ['Year', 'Month', 'Average Closing Price']
print(monthavgyear)
```

```python
# Find the highest monthly average for each year
himonthavgeachyr = monthavgyear.loc[monthavgyear.groupby('Year')['Average Closing Price'].idxmax()]
print(himonthavgeachyr)
```

```python
# Find the lowest monthly average for each year
lowmonthavgeachyr = monthavgyear.loc[monthavgyear.groupby('Year')['Average Closing Price'].idxmin()]
print(lowmonthavgeachyr)
```

```python
#make year and month a string
himonthavgeachyr['Year'] = himonthavgeachyr['Year'].astype(str)
himonthavgeachyr['Month'] = himonthavgeachyr['Month'].astype(str)

#make a plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Average Closing Price', hue='Month', dodge=False, data=himonthavgeachyr, palette='viridis')
plt.title('Highest Monthly Average Closing Prices per Year')
plt.xlabel('Year')
plt.show()
```

```python
#make year and month a string
lowmonthavgeachyr['Year'] = lowmonthavgeachyr['Year'].astype(str)
lowmonthavgeachyr['Month'] = lowmonthavgeachyr['Month'].astype(str)

#make a plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Average Closing Price', hue='Month', dodge=False, data=lowmonthavgeachyr, palette='viridis')
plt.title('Lowest Monthly Average Closing Prices per Year')
plt.xlabel('Year')
plt.show()
```
